## 5. 거스름돈 계산기


역할: 너는 파이썬 튜터야

목표: 아래 문제를 파이썬으로 풀어줘.

문제: 거스름돈 계산기를 만들어줘. 사용자가 1만원 단위의 금액을 입력하면, 상품값을 제외한 거스름돈은 5천원, 1천원, 500원, 100원의 거스름돈을 받을 수 있도록 해줘.

입력: (예) 지불금액  : 20000 상품금액 : 15400

출력: (예)==거스름돈 ==
5,000원권 : 0장 1,000원권 : 4장 500원 동전 : 0개  100원 동전 : 3개 (각 권종당 줄바꿈으로 처리)

제약: 몫과 나머지 연산자 사용, 1000원 이상은 지폐 500원, 100원은 동전으로 처리, 사용자가 10,000원 단위로 입력할 수 있도록 제시.

요청: 코드에 대해 한 줄씩 설명하고 테스트 3개 실행


```python
# TODO: 거스름돈 계산기

def calculate_change(paid_amount, item_price):
    """
    지불금액과 상품금액을 입력받아 거스름돈을 계산하고,
    5천원, 1천원 지폐와 500원, 100원 동전으로 분배하여 출력하는 함수입니다.
    """
    # 거스름돈 계산
    change = paid_amount - item_price

    # 거스름돈이 음수일 경우 처리
    if change < 0:
        print("지불 금액이 부족합니다.")
        return

    print("== 거스름돈 ==") # 거스름돈 출력 시작 알림

    # 5,000원권 계산
    num_5000 = change // 5000  # 몫 연산자로 5000원권 장수 계산
    change %= 5000              # 나머지 연산자로 남은 거스름돈 업데이트
    print(f"5,000원권 : {num_5000}장")

    # 1,000원권 계산
    num_1000 = change // 1000   # 몫 연산자로 1000원권 장수 계산
    change %= 1000              # 나머지 연산자로 남은 거스름돈 업데이트
    print(f"1,000원권 : {num_1000}장")

    # 500원 동전 계산
    num_500 = change // 500     # 몫 연산자로 500원 동전 개수 계산
    change %= 500               # 나머지 연산자로 남은 거스름돈 업데이트
    print(f"500원 동전 : {num_500}개")

    # 100원 동전 계산
    num_100 = change // 100     # 몫 연산자로 100원 동전 개수 계산
    change %= 100               # 나머지 연산자로 남은 거스름돈 업데이트
    print(f"100원 동전 : {num_100}개")

    # 100원 미만 잔돈이 있다면 추가 출력 (선택 사항)
    if change > 0:
        print(f"(남은 잔돈 : {change}원)")

# 사용자 입력 받기
# 지불금액은 만원 단위로 입력하도록 안내
paid_input = input("지불 금액을 입력하세요 (예: 10000, 20000 등 만원 단위): ")
item_price_input = input("상품 금액을 입력하세요: ")

try:
    paid_amount = int(paid_input)
    item_price = int(item_price_input)

    # 지불 금액이 만원 단위인지 확인 (선택 사항, 사용자에게만 안내)
    if paid_amount % 10000 != 0:
        print("주의: 지불 금액이 만원 단위가 아닙니다. 계산은 진행됩니다.")

    print(f"\n--- 사용자 입력 --- ")
    print(f"지불금액: {paid_amount}원, 상품금액: {item_price}원")
    calculate_change(paid_amount, item_price)
except ValueError:
    print("올바른 숫자를 입력해주세요.")

```

## 6. 성적 등급 계산기 (A/B/C/D/F + 범위 체크)


### 문제 : 점수를 입력받아 학점 계산
- 입력: 0~100 점수(score)
- 출력: A(100-90), B(89-80), C(79-70), D(69-60), F(그 외)
- 추가: 범위 밖 입력이면 `"입력 오류"` 출력


```python
# 점수를 입력받아 학점을 계산하는 프로그램

# 사용자에게 점수를 입력받습니다.
# 점수가 범위안에 있는지 확인합니다
score = int(input("점수를 입력하세요(0~100): "))

#  1) 점수가 정상 범위 밖이면 입력오류 출력
if score < 0 or score > 100:
    print("입력오류")

# 2)점수가 범위 안이면 등급계산

# 90점 이상 100점 이하이면 A
elif score >= 90:
    print("A")

# 80점 이상 89점 이하이면 B
elif score >= 80:
    print("B")

# 70점 이상 79점 이하이면 C
elif score >= 70:
    print("C")

# 60점 이상 69점 이하이면 D
elif score >= 60:
    print("D")

# 그 외, 즉 0점 이상 59점 이하이면 F
else:
    print("F")
```

```python
# 점수분석기
# 여러 개의 점수를 입력받아 리스트에 저장한 뒤
# 평균, 최대값, 최소값, 합계, 90점 이상 개수를 출력합니다.

# 사용자에게 점수를 여러 개 입력받습니다.
# 예: 80 95 70 100 60
score_input = input("점수를 공백으로 구분해서 입력하세요. 예: 80 95 70 100 60: ")

# 입력받은 문자열을 공백 기준으로 나눕니다.
# 예: "80 95 70" → ["80", "95", "70"]
score_text_list = score_input.split()

# 문자열 리스트를 정수 리스트로 변환합니다.
# 예: ["80", "95", "70"] → [80, 95, 70]
scores = []

for score_text in score_text_list:
    score = int(score_text)
    scores.append(score)

# 점수가 하나도 입력되지 않은 경우를 처리합니다.
if len(scores) == 0:
    print("입력오류: 점수를 하나 이상 입력해야 합니다.")

else:
    # 합계 계산
    total_sum_of_scores = 0
    for s in scores:
        total_sum_of_scores += s

    # 평균 계산
    average = total_sum_of_scores / len(scores)

    # 최대 점수 계산
    max_score = scores[0]
    for s in scores:
        if s > max_score:
            max_score = s

    # 최소 점수 계산
    min_score = scores[0]
    for s in scores:
        if s < min_score:
            min_score = s

    # 90점 이상인 점수 개수 세기
    count_90_or_more = 0
    for s in scores:
        if s >= 90:
            count_90_or_more += 1

    # 결과 출력
    print()
    print("===== 점수 분석 결과 =====")
    print("입력한 점수:", scores)
    print(f"합계: {total_sum_of_scores}점")
    print(f"평균: {average:.2f}점")
    print(f"최대: {max_score}점")
    print(f"최소: {min_score}점")
    print(f"90점 이상 개수: {count_90_or_more}개")
```


## 10. 숫자 맞히기 게임(while)


### 요구사항
- 정답 숫자를 맞힐 때까지 반복 입력
- 힌트: 작으면 UP, 크면 DOWN
- 추가: 시도 횟수 카운트
- 무한 루프 방지: 시도 횟수 제한(max_tries) 같은 안전장치를 두어도 좋습니다.


```python

# 숫자맞히기게임 테스트용 버전
# 테스트하기 쉽도록 정답을 37로 고정했습니다.

answer = 37
max_tries = 10
tries = 0

print("===== 숫자맞히기게임 =====")
print("1부터 100 사이의 숫자를 맞혀보세요.")
print(f"최대 {max_tries}번까지 시도할 수 있습니다.")
print()

while tries < max_tries:
    guess = int(input("숫자를 입력하세요: "))
    tries += 1

    if guess < answer:
        print("UP")

    elif guess > answer:
        print("DOWN")

    else:
        print()
        print("정답입니다!")
        print(f"총 {tries}번 만에 맞혔습니다.")
        break

    print(f"남은 시도 횟수: {max_tries - tries}번")
    print()

else:
    print()
    print("시도 횟수를 모두 사용했습니다.")
    print(f"정답은 {answer}였습니다.")
```

### 03.  강재 리스트 반복 출력 및 합계 [반복문]

문제: 리스트에 저장된 여러 개의 강재 무게 데이터를 반복문을 통해 하나씩 출력하고, 전체 강재의 총 무게를 계산하는 프로그램을 작성하세요.

```python
# 강재 무게 출력 및 총 무게 계산 프로그램
# 사용자가 여러 개의 강재 무게를 입력하면,
# 리스트에 저장한 뒤 하나씩 출력하고 총 무게를 계산합니다.

# 강재 무게를 공백으로 구분해서 입력받습니다.
# 예: 12.5 18.0 25.3 9.7 31.2
weight_input = input("강재 무게를 공백으로 구분해서 입력하세요(단위: 톤): ")

# 입력받은 문자열을 공백 기준으로 나눕니다.
# 예: "12.5 18.0 25.3" → ["12.5", "18.0", "25.3"]
weight_text_list = weight_input.split()

# 문자열 리스트를 실수형 리스트로 변환합니다.
# 예: ["12.5", "18.0", "25.3"] → [12.5, 18.0, 25.3]
steel_weights = []

for weight_text in weight_text_list:
    weight = float(weight_text)
    steel_weights.append(weight)

# 총 무게를 저장할 변수
total_weight = 0

print()
print("===== 강재 무게 목록 =====")

# 리스트에 저장된 강재 무게를 하나씩 출력
for index, weight in enumerate(steel_weights, start=1):
    print(f"{index}번 강재 무게: {weight}톤")

    # 총 무게에 현재 강재 무게를 더함
    total_weight += weight

print("========================")
print(f"전체 강재 총 무게: {total_weight}톤")
```
